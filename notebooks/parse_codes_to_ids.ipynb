```python
import json

with open("../data/trial_data/country_to_index.json", "r") as f:
    country_to_index = json.load(f)

country_to_index
```

```python
with open("../data/trial_data/country_code_to_name.json", "r") as f:
    country_code_to_name = json.load(f)

country_code_to_name
```

```python
country_code_to_index = {code: country_to_index[name] for code, name in country_code_to_name.items() if name in country_to_index}
country_code_to_index
```

```python
with open("../data/trial_data/country_code_to_index.json", "w") as f:
    json.dump(country_code_to_index, f)
```
